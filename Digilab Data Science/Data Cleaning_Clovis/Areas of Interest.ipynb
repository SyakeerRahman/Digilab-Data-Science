```python
import pandas as pd
import csv
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# string to store areas of interest
collection_of_interests = ""

with open('data/AFR-SG Member_Volunteer List (18 June 2022).csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    
#     collection_of_interests = " ".join(row[7].split(",") for row in csv_reader)
    
    for row in csv_reader:
        interests_column = row[7]
        
        # don't want to include title
        if interests_column != "What are some of your areas of interest(s)?":
                
            for interest in interests_column.split(","):
                collection_of_interests = collection_of_interests + " " + interest 
            
                
# print(collection_of_interests)


# Generate word cloud image
wordcloud = WordCloud(background_color='white').generate(collection_of_interests)
plt.figure(figsize=(40,20))

# Display the image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
        
```
