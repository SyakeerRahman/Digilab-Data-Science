# Wrangling and Cleaning
__Cleaning needs__

1. Change the header name to easy to read and code.
2. Source change to int.
3. For header "Age group / Year of Birth" use range and put in A, B, C.
4. For the last four header take the important word as the main, or put in different category because it looks like a multiple choice, so make sure to change to other coloumn based on choice or you can group based on choice for easy reading, manipulation, and modelling.
5. No need to delete null/NaN data because the data to small to ignore the rest of column.

#### Rename the header

```python
rawdata.head()
```

```python
# Rename the header
rawdata1 = rawdata.rename(index=str, columns={"Age group / Year of Birth": "Age",
                                              "Nationality (please indicate)": "Nationality",
                                              "Occupation/ Job title" : "Job",
                                              "Why did you join AFR-SG? / \nHow do you wish to contribute to the cause?": "Reason/Contribute",
                                              "What are some of your areas of interest(s)?": "Interest",
                                              "Do you have any skills, experiences, strengths that you could offer to the group? (please only choose what is relevant to you)": "Skills/Experiences/Strengths",
                                              "How much time do you have available for volunteering? (e.g. 1 - 2 hours a week; weekdays; evenings; weekends; shifts; flexible)": "Available Time"})
rawdata1.head()
```

#### Change the data types

```python
# use the rawdata1
rawdata1["Source"]=rawdata1["Source"].astype(int)
```

```python
rawdata1.dtypes
```

```python
rawdata1.head()
```

#### Change The Timestamp to Date and Time

```python
rawdata1.head()
```

```python
rawdata1['Date'] = [d.date() for d in rawdata1['Timestamp']]
rawdata1['Time'] = [d.time() for d in rawdata1['Timestamp']]
```

```python
rawdata1.head()
```

```python
new_cols = ["Source", "Timestamp", "Date", "Time", "Gender", "Age", "Nationality", "Job", "Reason/Contribute", "Interest",
            "Skills/Experiences/Strengths", "Available Time"]
rawdata1=rawdata1[new_cols]
```

```python
rawdata1.head()
```

```python
# Delete Timestamp and time
rawdata1 = rawdata1.drop(rawdata1.columns[[1, 3]], axis=1)
rawdata1.head()
```

#### Change Gender to use 1, 2 or 3

```python
rawdata1.head()
```

```python
rawdata1.loc[rawdata1["Gender"]=="Male", ["Gender"]]=1
rawdata1.loc[rawdata1["Gender"]=="Female", ["Gender"]]=2
rawdata1.loc[rawdata1["Gender"]=="Prefer not to disclose", ["Gender"]]=3
rawdata1.loc[rawdata1["Gender"]=="", ["Gender"]]=4
```

#### Change Age Header to use same format

```python
rawdata1["Age"].value_counts()
```

__The data  have different format__, so change the same format. Change the year to range so it will made easier to cluster

```python
rawdata1.head()
```

```python
# Change the range that have space
rawdata1.loc[rawdata1['Age']=="19 - 24", ['Age']]="19-24"
rawdata1.loc[rawdata1['Age']=="25 - 34", ['Age']]="25-34"
rawdata1.loc[rawdata1['Age']=="35 - 44", ['Age']]="35-44"
rawdata1.loc[rawdata1['Age']=="45 - 54", ['Age']]="45-54 "
```

```python
rawdata1["Age"].value_counts()
```

```python
# Change all data to range
rawdata1.loc[rawdata1["Age"]=="> 18", ["Age"]]=">=18"
rawdata1.loc[rawdata1["Age"]==2004.0, ["Age"]]=">=18"
rawdata1.loc[rawdata1["Age"]==2005.0, ["Age"]]="<18"
rawdata1.loc[rawdata1["Age"]=="< 55", ["Age"]]="45-54"
rawdata1.loc[rawdata1["Age"]=="<55", ["Age"]]="45-54"
rawdata1.loc[rawdata1["Age"]==1968.0, ["Age"]]="45-54"
rawdata1.loc[rawdata1["Age"]=="1/8/1999", ["Age"]]="19-24"
rawdata1.loc[rawdata1["Age"]==1012000.0, ["Age"]]="19-24"
rawdata1.loc[rawdata1["Age"]==1965.0, ["Age"]]=">55"
rawdata1.loc[rawdata1["Age"]=="01_01_2000", ["Age"]]="<18"
```

```python
rawdata1["Age"].value_counts()
```

```python
# Change the datatypes to float
for a in rawdata1.index:
    try:
        rawdata1.loc[a, "Age group / Year of Birth"]=float(rawdata1["Age group / Year of Birth"][a])
    except:
        pass
```

```python
# Find all the data in columns age that have types float
age_1018 = [i for i in range(10,18)]
age_1924 = [i for i in range(19,25)]
age_2534 = [i for i in range(25,35)]
age_3544 = [i for i in range(35,45)]
age_4554 = [i for i in range(45,54)]
age_5560 = [i for i in range(55,60)]
for a in rawdata1.index:
    if type(rawdata1["Age"][a])==float:
        if (2022.0-rawdata1["Age"][a] in age_1018):
            rawdata1.loc[a, ["Age"]]="10-18"
            continue
        if (2022.0-rawdata1["Age"][a] in age_1924):
            rawdata1.loc[a, ["Age"]]="19-24"
            continue
        if (2022.0-rawdata1["Age"][a] in age_2534):
            rawdata1.loc[a, ["Age"]]="25-34"
            continue
        if (2022.0-rawdata1["Age"][a] in age_3544):
            rawdata1.loc[a, ["Age"]]="35-44"
            continue
        if (2022.0-rawdata1["Age"][a] in age_4554):
            rawdata1.loc[a, ["Age"]]="45-54"
            continue
        if (2022.0-rawdata1["Age"][a] in age_5560):
            rawdata1.loc[a, ["Age"]]="55-60"
            continue
```
